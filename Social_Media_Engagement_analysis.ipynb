```python
df_clean['Timestamp'] = pd.to_datetime(df_clean['Timestamp'])
```

```python
df_clean['Date'] = df_clean['Timestamp'].dt.date
df_clean['Month'] = df_clean['Timestamp'].dt.month_name()
df_clean['Year'] = df_clean['Timestamp'].dt.year
```

```python
for col in ['Platform', 'Content_Type', 'Category', 'Sentiment']:
    print(f"\n{col}")
    print(df_clean[col].unique())
```

```python
text_columns = ['Platform', 'Content_Type', 'Category', 'Sentiment']

for col in text_columns:
    df_clean[col] = df_clean[col].str.strip().str.title()
```

```python
#Feature engineering
df_clean['Total_Engagement'] = (
    df_clean['Likes'] +
    df_clean['Comments'] +
    df_clean['Shares'] +
    df_clean['Saves']
)
```

```python
#Engagement per view
df_clean['Engagement_Per_View'] = (
    df_clean['Total_Engagement'] /
    df_clean['Views']
).round(4)
```

```python
#Weekend or Weekday
weekend = ['Saturday', 'Sunday']

df_clean['Post_Type'] = np.where(
    df_clean['Day_of_Week'].isin(weekend),
    'Weekend',
    'Weekday'
)
```

```python
#Hashtag Category
def hashtag_category(x):
    if x <= 3:
        return 'Low'
    elif x <= 7:
        return 'Medium'
    else:
        return 'High'

df_clean['Hashtag_Category'] = df_clean['Hashtag_Count'].apply(hashtag_category)
```

```python
'''Insights from the data analysis:
1. Platform-wise Engagement
a) Observation:  TikTok recorded the highest average total engagement (approximately 300,000), 
while LinkedIn had the lowest average engagement (below 5,000).
b) Business Insight: Short-form video platforms such as TikTok encourage significantly higher user interaction
compared to professional networking platforms like LinkedIn.
c) Recommendation: Increase investment in short-form video content for platforms with high engagement while
tailoring professional content specifically for LinkedIn.

2. Category-wise Engagement  
a) Observation: Food, Entertainment, Education, and Travel categories generated the highest engagement,
whereas Business and Fashion received comparatively lower engagement.
b) Business Insight: Users are more likely to interact with informative and entertaining content than with 
business-focused posts.
c) Recommendation: Prioritize high-performing categories while experimenting with new strategies to improve engagement in 
lower-performing categories.

3. Best Posting day 
a) Observation: Wednesday recorded the highest average engagement among all days of the week.
b) Business Insight: Mid-week posts appear to capture greater audience attention and interaction.
c) Recommendation: Schedule important campaigns and promotional posts on Wednesdays whenever possible.

4. Best Posting Hour
a) Observation: Engagement peaked around 7:00 PM (19:00 hours) and was lowest around 9:00 AM.
b) Business Insight: Users are more active on social media during the evening than in the morning.
c) Recommendation: Schedule high-priority posts during evening hours to maximize engagement.

5. Sentiment Analysis
a) Observation: Posts with positive sentiment achieved the highest engagement, followed by neutral and negative sentiment.\
b) Business Insight: Positive content tends to encourage greater audience interaction.
c) Recommendation: Maintain a positive and engaging tone in content wherever appropriate.

6. Comments -> Total engagement
a) Observation : Posts receiving more comments generally achieve higher engagement.
b) Business Insight : Audience conversations significantly increase engagement.
c) Recommendation : Ask questions, conduct polls, and encourage discussions in captions.

7. Shares ->  Total Engagement
a) Observation : Sharing behavior contributes substantially to total engagement.
b) Business Insight : Highly shareable content tends to perform better.
c) Recommendation : Publish informative, entertaining, or inspirational content that users want to share.

8. Views ↔ Saves
a) Observation : Posts with more views are also more likely to be saved.
b) Business Insight : Users tend to save content they find useful after viewing it.
c) Recommendation : Create educational posts, guides, tips, and tutorials that users may want to revisit.

9. Follower Count
a) Observation : Having more followers does not necessarily result in higher engagement.
b) Business Insight : Audience size alone is not enough; content quality and relevance matter more.
c) Recommendation : Focus on producing engaging content rather than only increasing follower count.

10. Engagement Rate ↔ Follower Count
a) Observation : As follower count increases, engagement rate tends to decrease slightly.
b) Business Insight : Larger accounts often experience lower engagement rates because reaching and engaging every follower becomes more difficult.
c) Recommendation : Large creators should prioritize maintaining audience interaction instead of focusing only on follower growth.

Future Scope

• Analyze engagement trends over multiple years.
• Develop predictive models to forecast engagement.
• Perform sentiment analysis using Natural Language Processing (NLP).
• Build an interactive Power BI dashboard for real-time monitoring.
• Compare platform performance across different regions and demographics.
'''
```
